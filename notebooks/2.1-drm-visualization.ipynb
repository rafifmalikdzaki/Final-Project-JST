```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/comb_evaluation.csv", index_col=0)
df
```

```python
for col in df.columns[1:-1]:
    df.groupby(["dMethod",col])["Akurasi"].max().plot(kind="bar", ylabel="Akurasi",ylim=(.3, .6), title=col)
    plt.show()
```

```python
df.groupby(["dMethod"])["Akurasi"].max().plot(kind="bar", ylabel="Akurasi",ylim=(.3, .6))
plt.show()
```
